```python
#boxplot
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
sns.boxplot(dados['Anos de Experiencia'])
plt.subplot(1,2,2)
sns.boxplot(dados['Renda'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Analisa da Renda')
sns.kdeplot(dados['Renda'], shade=True);
```

```python
plt.figure(figsize=(10,5))
plt.title('Tempo de experiencia')
sns.kdeplot(dados['Anos de Experiencia'], shade=True);
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'Anos de Experiencia', y = 'Renda', data = dados);
```

```python
#traçar a linha de regressão com intervalo de confiança
plt.figure(figsize = (12,6))
sns.regplot(data = dados, x = 'Anos de Experiencia', y = 'Renda', ci = 95);
```

```python
#correlacao de pearson
cor = dados.corr('pearson')
sns.heatmap(cor, annot = True, cmap = 'Blues');
```

```python
#correlacao de spearman
cor = dados.corr('spearman')
sns.heatmap(cor, annot = True, cmap = 'Blues');
```
